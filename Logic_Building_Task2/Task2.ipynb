Problem 1 — Unique Words in a Sentence

```python

sentence = input("Enter sentence: ")
words = sentence.split()

unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

Problem 2 — Highest Salary from Employee Data

```python

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)

print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

Problem 3 — Find Maximum and Minimum Values

```python

numbers = [45, 22, 89, 10, 66]

max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

Problem 5 — Calculate Attendance Percentage

```python

attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

Problem 6 — Remove Duplicate Phone Numbers

```python

phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

Problem 7 — Count Character Frequency

```python

text = "python"
frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
```

Problem 9 — Check if Key Exists in Dictionary

```python

employees = {"Ravi": 75000, "Anita": 68000}

key_to_check = "Ravi"

if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

Problem 10 — Calculate Average Marks

```python

marks = [70, 75, 80, 65]

average = sum(marks) / len(marks)

print("Average Marks:", average)
```
