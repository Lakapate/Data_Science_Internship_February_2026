1. User Login Check

```python
username = "admin"
password = "1234"
print("Enter the username and password")
x = input("Username")
y = input("Password")
if((x==username) and (y==password)):
    print("Login Successful")
else:
    print("Invalid Credentials")
```

2. Pass / Fail Analyzer

```python
marks = [45,78,90,33,60]
pass_students = 0
fail_students = 0
for i in marks:
    if(i>=50):
        pass_students = pass_students+1
    else:
        fail_students = fail_students+1

print(f"Number of pass students are = {pass_students}")
print(f"Number of fail students are = {fail_students}")
```

3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []

for name in names:
    cleaned_names.append(name.strip().lower())
print(cleaned_names)
```

4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length = len(msg)
    print("Message:",msg)
    print("Length:",length)
    
    if(length>10):
     print("FLAGGED (to long)")
    print()

```

5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0

for text in logs:
    if(text == "ERROR"):
        error_count = error_count+1
print(f"Number of error = {error_count}")
```
